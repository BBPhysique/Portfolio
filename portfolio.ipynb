### A- Visualisation d'ondes ultrasonores par Schlieren

<details>
<summary>En savoir plus</summary>
En introduction du chapitre d'acoustique, observation d'une onde ultrasonore produite par un lévitateur acoustique (les deux tâches noires dédoublées sont deux morceaux de polystyrène piégés). 
    
La LED servant de source est alimentée par des impulsions répétées à quasiment 40 kHz, provoquant par moment un effet stroboscopique. 
    
En classe l'expérience débute par l'observation du gaz d'un briquet pour donner un sens à chaque couleur et relier l'observation à l'effet mirage faisant sortir un faisceau lumineux du filtre lorsqu'un gradient de température existe proche du miroir.

### D- Conception 3D : tuyère pour effet Venturi et support pour piles et électrolyseurs

<details>
<summary>En savoir plus</summary>


Conception 3D d'une tuyère servant à démontrer l'effet Venturi, lorsque de l'eau d'une pompe de bassin 3000L/h alimente la tuyère. Après démonstration de la relation en classe, scission de la classe en 4 sous-groupes chargés de mesurer indépendamment $P_A$, $P_B$, $v_A$, et $v_B$, puis vérification expérimentale de la loi démontrée juste avant. Le 3ème tuyau percé n'est pas au contact du fluide si la sortie est ouverte et non contrainte. 

Un tuyau fixé au premier trou permet de fixer un tube de 1mm de diamètre pour mesure de la pression hydrostatique, le tuyau central est en pratique en dépression : lorsque non bouché des bulles apparaissent spontanément en sortie par aspiration d'airau tuyau central. L'hypothèse de cavitation est écartée en bouchant le-dit tuyau central avec un doigt, faisant disparaître les bulles.

![Tuyère maison](tuyere.png "Tuyère maison")

Pièces imprimées, à droite un support pour électrolyse (Guides pour cylindres de graphites en bas, guides pour tubes à essai en haut, guides pour plaques de métal en haut). Le guide pour tube à essai est troué en bas afin que les bulles de gaz puissent remonter pour être capturées. Le diamètre total du support est pensé pour tenir dans un bécher de 200 mL.

![Tuyère et support d'électrolyse](3D.jpg "Tuyère et support d'électrolyse")

</details>

## II- Créations expérimentales

### A- Petite chambre à brouillard

<details>
<summary>En savoir plus</summary>
Pour démonstration en Enseignement Scientifique de la notion de désintégration radioactive, construction d'une petite chambre à brouillard. Après plusieurs prototypes, le refroidissement à eau via une pompe d'aquarium dans un bain de glace est la solution la plus performante pour refroidir les modules Peltier. Telle que présentée aux élèves, dans la démonstration vidéo le circuit haute tension est volontairement mis de côté, et aucune source radioactive n'est introduite. Le mouchoir est imbibé d'isopropanol.


NB : La vidéo est une illustration du montage et non un support pédagogique destiné aux élèves.

[![Lien vidéo](https://img.youtube.com/vi/p4-ajnBrQ-g/maxresdefault.jpg)](https://www.youtube.com/watch?v=p4-ajnBrQ-g)


</details>



### B- Mesure de la vitesse de la lumière par temps de vol sur 2m de distance

<details>
<summary>En savoir plus</summary>
Pour illustration de la vitesse de la lumière auprès d'élèves en 2nde, illustration du temps de vol de la lumière sur 2 mètres. Le circuit de la diode laser est alimenté à 1MHz par des impulsions de temps de montée d'environ 5 ns. Le repère de lecture du temps de réception de l'impulsion par la photodiode rapide est fait à la demi-amplitude montante pour une meilleur sensibilité. Erreur relative d'environ 10% sur la mesure de c. Le montage électrique a été fait dans la limite de mes outils et le montage optique dans la limite des outils de mon lycée.

NB : la vidéo a été tournée à destination d'un ami, il s'agit d'une illustration du montage et non un support pédagogique destiné aux élèves.

![Montage c](c2m.png "Montage c")

[![Lien vidéo](https://img.youtube.com/vi/M_8xNJQ3Hyk/maxresdefault.jpg)](https://www.youtube.com/watch?v=M_8xNJQ3Hyk)
