## II- Créations expérimentales

### A- Petite chambre à brouillard

<details>
<summary>En savoir plus</summary>
Pour démonstration en Enseignement Scientifique de la notion de désintégration radioactive, construction d'une petite chambre à brouillard. Après plusieurs prototypes, le refroidissement à eau via une pompe d'aquarium dans un bain de glace est la solution la plus performante pour refroidir les modules Peltier. Telle que présentée aux élèves, dans la démonstration vidéo le circuit haute tension est volontairement mis de côté, et aucune source radioactive n'est introduite. Le mouchoir est imbibé d'isopropanol.


NB : La vidéo est une illustration du montage et non un support pédagogique destiné aux élèves.
[![Lien vidéo](https://youtu.be/p4-ajnBrQ-g)](https://youtu.be/p4-ajnBrQ-g)

</details>

### B- Modulateur Spatial de Lumière fait maison

<details>
<summary>En savoir plus</summary>
Avec un ancien vidéoprojecteur tri-LCD, réalisation d'un SLM à l'aide d'un des petits écrans LCD. Le vidéoprojecteur apparaît comme un vidéoprojecteur à n'importe quel ordinateur y étant connecté. Les élèves peuvent donc vidéoprojeter n'importe quelle diapositive de leur conception depuis l'ordinateur, et observer de façon dynamique la transformée de Fourier optique de cet objet.  

IMG SLM
</details>



### C- Mesure de la vitesse de la lumière par temps de vol sur 2m de distance

<details>
<summary>En savoir plus</summary>
Pour illustration de la vitesse de la lumière auprès d'élèves en 2nde, illustration du temps de vol de la lumière sur 2 mètres. Le circuit de la diode laser est alimenté à 1MHz par des impulsions de temps de montée d'environ 5 ns. Le repère de lecture du temps de réception de l'impulsion par la photodiode rapide est fait à la demi-amplitude montante pour une meilleur sensibilité. Erreur relative d'environ 10% sur la mesure de c. Le montage électrique a été fait dans la limite de mes outils et le montage optique dans la limite des outils de mon lycée.

NB : la vidéo a été tournée à destination d'un ami, il s'agit d'une illustration du montage et non un support pédagogique destiné aux élèves.

[![Lien vidéo](https://youtu.be/M_8xNJQ3Hyk)](https://youtu.be/M_8xNJQ3Hyk)

</details>
